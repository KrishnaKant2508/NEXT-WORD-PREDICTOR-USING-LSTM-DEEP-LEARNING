```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer(oov_token = '<oov>')
data['Description'] = data['Description'].astype(str)
tokenizer.fit_on_texts(data['Description'])
word_index_length = len(tokenizer.word_index)
print(word_index_length)
print("dpr: ", tokenizer.word_index['dpr'])
```

```python
tokenizer.word_index
```

```python
input_sequence =[]
for sentence in data['Description']:
  token_list = tokenizer.texts_to_sequences([sentence])[0]
  # print(sentence)
  # print(token_list)
  for i in range(1,len(token_list)):
    input_sequence.append(token_list[:i+1])
input_sequence

```

```python
len(input_sequence)
```

```python
input_seq_list = []
for x in input_sequence:
    input_seq_list.append(len(x))
max_len = max(input_seq_list)
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequence, maxlen = max_len, padding='pre')
padded_input_sequences
```

```python
x = padded_input_sequences[:,:-1]
y = padded_input_sequences[:,-1]
x.shape,y.shape
```

```python
from tensorflow.keras.utils import to_categorical
ys = to_categorical(y,num_classes=word_index_length+1)
ys
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM, Dense
```

```python
model = Sequential()
model.add(Embedding(2861,100,input_length=238))
model.add(LSTM(150))
model.add(Dense(2861,activation = 'softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer = 'adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x,ys,epochs=3)
```

```python
import time
text = "not able to submit "
def word_predictor(text):
  for i in range(20):
    # tokenize
    token_text = tokenizer.texts_to_sequences([text])[0]
    # padding
    padded_token_text = pad_sequences([token_text], maxlen=238, padding='pre')
    # predict
    pos = np.argmax(model.predict(padded_token_text))

    for word,index in tokenizer.word_index.items():
      if index == pos:
        text = text + " " + word
        return text
        time.sleep(1)
word_predictor(text)
```
